```python
import pandas as pd
```

```python
raw_data: pd.DataFrame = pd.read_excel("rawData.xlsx")
raw_data.info()
```

```python
raw_data['CTO NUMBER'] = raw_data['CTO NUMBER'].str.replace("-2025", "")
raw_data['ROLE'] = raw_data['ROLE'].str.strip().str.upper()
raw_data['CTO CLASS'] = raw_data['CTO CLASS'].str.strip().str.upper()
raw_data['CTO ADDED VALUE'] = raw_data['CTO ADDED VALUE'].fillna(0).astype('int64')
raw_data['EMPLOYEE DNI'] = raw_data['EMPLOYEE DNI'].astype(str)
raw_data['DNI SOURCE'] = raw_data['DNI SOURCE'].str.strip().str.upper()
```

```python
raw_data.columns = (raw_data.columns.str.strip().str.replace(" ", "_").str.lower())
```

```python
raw_data.info()
raw_data.to_csv("finalData.csv", index=False)
```
